#### Check missing values

```python
df.isnull().sum()
```

 Few columns in the dataset have missing values. These columns require appropriate data handling before further analysis.

#### Fill missing values

```python
df['director']=df['director'].fillna('Unknown')
df['cast']=df['cast'].fillna('Unknown')
df['country']=df['country'].fillna('Unknown')
```

```python
df.isnull().sum()
```

#### Drop rows with missing date_added

```python
df=df.dropna(subset=['date_added'])
```

 The 10 rows with missing values in the date_added column were removed because the missing fates could not be reliable inputed, ensuring better quality for analysis.

```python
# Fill missing rating and duration with mode
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
```

```python
df.isnull().sum()
```

#### Remove Duplicate Records

```python
#check for duplicate rows
df.duplicated().sum()
```

No duplicate records found.

#### Standardize text values

```python
df['director']=df['director'].str.strip()
df['cast']=df['cast'].str.strip()
df['country']=df['country'].str.strip()
df['type']=df['type'].str.strip()
df['rating']=df['rating'].str.strip()
```

#### Convert Date Format

```python
df['date_added'].dtype
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
```

```python
df['date_added'].dtype
```

```python
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
df['date_added'].head(10)
```

#### Rename Column Headers

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

#### Check and Fix Data Types

```python
df.dtypes
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%d-%m-%Y')
```

Verified the data types of all columns and corrected them where necessary to ensure consistency and support accurate data analysis.
